```python
import pandas as pd
data_1 = pd.read_csv('Student_performance_data _.csv')
y=data_1['GradeClass']
data_1.drop(['StudentID', 'GradeClass'], axis=1, inplace=True)
print(data_1.head())
print(y.head())
```

## train the model

```python
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix, precision_score, recall_score
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test=train_test_split(data_1, y, train_size=0.8)

```

```python
model = DecisionTreeClassifier()
model.fit(x_train,y_train)
```

```python
model_1=SVC()
model_1.fit(x_train, y_train)
```

```python
y_pred_dt=model.predict(x_test)
y_pred_sv=model_1.predict(x_test)

```

```python
def evaluate_model(y_true, y_pred, model_name):
    print(f"--- {model_name} ---")
    print("Accuracy:", accuracy_score(y_true, y_pred))
    print("Precision:", precision_score(y_true, y_pred, average='weighted'))
    print("Recall:", recall_score(y_true, y_pred, average='weighted'))
    print("F1-Score:", f1_score(y_true, y_pred, average='weighted'))
    print("Confusion Matrix:\n", confusion_matrix(y_true, y_pred))

evaluate_model(y_test, y_pred_dt, "Decision Tree")
evaluate_model(y_test, y_pred_sv, "SVM")
```

```python
import pickle

# Save Decision Tree model
with open('decision_tree_model.pkl', 'wb') as f:
    pickle.dump(model, f)

# Save Random Forest model
with open('svm_model.pkl', 'wb') as f:
    pickle.dump(model_1, f)
```
